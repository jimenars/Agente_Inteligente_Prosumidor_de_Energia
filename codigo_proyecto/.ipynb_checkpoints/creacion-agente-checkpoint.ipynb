```python
# Crear DataFrame con índice de tiempo
horas = pd.date_range(start='2024-04-12 00:00:00', end='2024-04-13 23:59:59', freq='s')
df = pd.DataFrame(index=horas)
df['Potencia total'] = 0

# Iterar sobre los timeslots y sumar la potencia para cada intervalo de tiempo
for slot in timeslots:
    tiempo_inicial = pd.Timestamp(slot['Start'])
    tiempo_final = pd.Timestamp(slot['End']) if 'End' in slot else tiempo_inicial
    power = slot['Power']
    df.loc[tiempo_inicial:tiempo_final, 'Potencia total'] += power
    
# Leer el archivo CSV de producción
pre_optimizacion = pd.read_csv(pasos_tiempo_minutos + '/netload.csv', sep=';')
pre_optimizacion.columns = ['Date', 'Demand', 'PV_Production', 'Wind_Production', 'Production', 'Netload']
pre_optimizacion['Date'] = pd.to_datetime(pre_optimizacion['Date'])
pre_optimizacion.set_index('Date', inplace=True)


# Crear la figura y los ejes (con grafica de producción)
fig, ax1 = plt.subplots(figsize=(10, 6))

# Graficar el consumo total de energía en el primer eje y
ax1.plot(df.index, df['Potencia total'], color='b')
ax1.set_xlabel('Hora del día')
ax1.set_ylabel('Potencia total (W)', color='b')
ax1.tick_params('y', colors='b')

# Crear un segundo eje y compartido para la producción
ax2 = ax1.twinx()
ax2.plot(pre_optimizacion.index, pre_optimizacion["Production"], color='r', linestyle='--')
ax2.set_ylabel('Producción (W)', color='r')
ax2.tick_params('y', colors='r')

plt.title('Consumo total de energía y producción a lo largo del día')
plt.xticks(rotation=90)
plt.show()
```

```python
# Establecer la semilla para reproducibilidad
random.seed(42)

flexibilidad_dispositivos = {}

# Generar números aleatorios únicos para cada electrodoméstico
flexibilidades_unicas = random.sample(range(1, 7), len(dispositivos_flexibles))
for dispositivo, flexibilidad in zip(dispositivos_flexibles, flexibilidades_unicas):
    flexibilidad_dispositivos[dispositivo] = flexibilidad

print("Diccionario de flexibilidad aleatoria con la semilla fijada:")
print(flexibilidad_dispositivos)

# Nombre del archivo CSV donde se guardará el diccionario
nombre_archivo_csv = "datos/flexibilidad_dispositivos.csv"

# Escribir el diccionario en el archivo CSV
with open(nombre_archivo_csv, 'w', newline='') as archivo_csv:
    nombres_campos = ['Dispositivos', 'Flexibilidad']
    escritor = csv.DictWriter(archivo_csv, fieldnames=nombres_campos)

    escritor.writeheader()
    for dispositivo, flexibilidad in flexibilidad_dispositivos.items():
        escritor.writerow({'Dispositivos': dispositivo, 'Flexibilidad': flexibilidad})

print("El diccionario de flexibilidad se ha guardado en el archivo CSV:", nombre_archivo_csv)

num_datos = 17
```

```python
# Obtener las fechas de inicio y fin de los timeslots
fecha_inicio = pd.Timestamp(timeslots[0]['Start']).floor('D')
fecha_fin = fecha_inicio + pd.Timedelta(days=1)

# Crear un DataFrame con índice de tiempo para cada minuto del día utilizando las fechas de los timeslots
minutos_del_dia = pd.date_range(start=fecha_inicio, end=fecha_fin, freq='T')
df = pd.DataFrame(index=minutos_del_dia)

# Crear un diccionario para almacenar los números de "dispositivo" por fecha
numeros_dispositivos_por_hora_de_inicio = {}

# Iterar sobre los timeslots y asignar el dispositivo a cada minuto en el DataFrame
for slot in timeslots:
    hora_inicio = pd.Timestamp(slot['Start'])
    dispositivo = slot['Appliance']
    
    # Obtener el número correspondiente al dispositivo
    numero_dispositivo = flexibilidad_dispositivos.get(dispositivo)
    
    # Reemplazar None por 0
    numero_dispositivo = 0 if numero_dispositivo is None else numero_dispositivo

    # Verificar si ya existe una lista para este tiempo de inicio
    if hora_inicio in numeros_dispositivos_por_hora_de_inicio:
        # Si ya existe, agregamos el nuevo número de dispositivo a la lista
        numeros_dispositivos_por_hora_de_inicio[hora_inicio].append(numero_dispositivo)
    else:
        # Si no existe, creamos una nueva lista con el número de dispositivo
        numeros_dispositivos_por_hora_de_inicio[hora_inicio] = [numero_dispositivo]

# Obtener las fechas únicas del diccionario
fechas = sorted(set(fecha.date() for fecha in numeros_dispositivos_por_hora_de_inicio.keys()))

# Crear un DataFrame con las filas y columnas apropiadas
df = pd.DataFrame(index=numeros_dispositivos_por_hora_de_inicio.keys(), columns=fechas)

# Llenar el DataFrame con los valores del diccionario
for timestamp, valores in numeros_dispositivos_por_hora_de_inicio.items():
    fecha = timestamp.date()
    df.loc[timestamp, fecha] = valores

# Llenar los valores NaN con una lista vacía
df.fillna("0", inplace=True)
df.fillna(value={}, inplace=True)

# Convertir el índice a formato de hora
df.index = df.index.strftime('%H:%M:%S')

import numpy as np

# Definir una función de reemplazo personalizada
def reemplazo_personalizado(valor):
    if isinstance(valor, list):
        # Filtrar los elementos distintos de cero en la lista
        valores_filtrados = [v for v in valor if v != 0]
        # Verificar si hay elementos distintos de cero en la lista filtrada
        if valores_filtrados:
            # Devolver la lista filtrada
            return valores_filtrados
        else:
            # Si la lista filtrada está vacía, devolver 0
            return 0
    else:
        return valor

# Aplicar la función de reemplazo personalizada a cada elemento del DataFrame
df = df.applymap(reemplazo_personalizado)

# Mostrar el DataFrame
display(df)

# Expandir las listas en el DataFrame
valores_expandidos = df[df != '0'].stack().explode()

# Reemplazar las listas [0] con 0
valores_expandidos = valores_expandidos.apply(lambda x: 0 if x == [0] else x)

# Contar los valores distintos de 0
total_de_conteos_no_cero = (valores_expandidos[valores_expandidos != 0].value_counts().sum())

# Mostrar el total de conteos
print(total_de_conteos_no_cero)
```

```python
#Ver cuantos datos cojo 
if (total_de_conteos_no_cero < num_datos ):
    num_datos = total_de_conteos_no_cero
    
print(num_datos)
```

```python
# Crear una lista para almacenar los números de dispositivos de cada fila
numeros_dispositivos_lista = []

# Iterar sobre las filas del DataFrame
for indice, fila in df.iterrows():
    # Crear una lista para almacenar los números de dispositivos de la fila actual
    numeros_dispositivos = []
    
    # Iterar sobre los valores de la fila actual
    for valor in fila:
        # Verificar si el valor no es "0" y no es NaN
        if valor != "0":
            if isinstance(valor, list):
                # Extender la lista de números de dispositivos con los valores de la lista
                numeros_dispositivos.extend(valor)
    
    # Dividir la lista de números de dispositivos en sublistas de un solo elemento
    for numero_dispositivo in numeros_dispositivos:  # Excluir el último elemento
        numeros_dispositivos_lista.append([numero_dispositivo])


# Mostrar la lista generada
print(numeros_dispositivos_lista)

# Seleccionar los primeros X elementos de la lista
numeros_dispositivos_lista_seleccionados = numeros_dispositivos_lista[:num_datos-1]

# Mostrar la lista seleccionada
print(numeros_dispositivos_lista_seleccionados)

# Convertir la lista seleccionada a un DataFrame de pandas
df_numeros_dispositivos_seleccionados = pd.DataFrame(numeros_dispositivos_lista_seleccionados)

# Guardar el DataFrame como un archivo CSV sin encabezado
df_numeros_dispositivos_seleccionados.to_csv('datos/flexibilities.csv', header=False, index=False)
```

```python
# Crear un diccionario para almacenar los números de "número de electrodoméstico" por fecha
diccionario_fecha = {}

# Iterar sobre los timeslots y asignar el electrodoméstico a cada minuto en el DataFrame
for intervalo in timeslots:
    hora_inicio = pd.Timestamp(intervalo['Start'])
    electrodomestico = intervalo['Appliance']
    
    # Verificar si ya existe una lista para este tiempo de inicio
    if hora_inicio in diccionario_fecha:
        # Si ya existe, agregamos el nuevo número de electrodoméstico a la lista
        diccionario_fecha[hora_inicio].append(hora_inicio)
    else:
        # Si no existe, creamos una nueva lista con el número de electrodoméstico
        diccionario_fecha[hora_inicio] = [hora_inicio]

# Obtener las fechas únicas del diccionario
fechas = sorted(set(fecha.date() for fecha in diccionario_fecha.keys()))

# Crear un DataFrame con las filas y columnas apropiadas
df = pd.DataFrame(index=diccionario_fecha.keys(), columns=fechas)

# Llenar el DataFrame con los valores del diccionario
for timestamp, valores in diccionario_fecha.items():
    fecha = timestamp.date()
    df.loc[timestamp, fecha] = valores

# Llenar los valores NaN con una lista vacía
df.fillna("0", inplace=True)

# Convertir el índice a formato de hora
df.index = df.index.strftime('%H:%M:%S')

# Mostrar el DataFrame
# display(df)

fechas_lista = []

# Iterar sobre las filas del DataFrame
for indice, fila in df.iterrows():
    # Crear una lista para almacenar las fechas de electrodomésticos de la fila actual
    fechas_electrodomesticos = []
    
    # Iterar sobre los valores de la fila actual
    for valor in fila:
        # Verificar si el valor no es "0" y no es NaN
        if valor != "0":
            if isinstance(valor, list):
                # Convertir la hora a formato HH.MM
                sublist_formateada = pd.to_datetime(valor).strftime('%H.%M')
                fechas_electrodomesticos.extend(sublist_formateada)
    
    # Dividir la lista de fechas de electrodomésticos en sublistas de un solo elemento
    for fecha_electrodomestico in fechas_electrodomesticos:
        fechas_lista.append([fecha_electrodomestico])        

# Mostrar la lista generada
print(fechas_lista)

fechas_lista_seleccionadas = fechas_lista[:num_datos]

print(fechas_lista_seleccionadas)

# Convertir la lista de listas a un DataFrame de pandas
df_fechas = pd.DataFrame(fechas_lista_seleccionadas)

# Guardar el DataFrame como un archivo CSV sin el encabezado
df_fechas.to_csv('datos/dates.csv', header=False, index=False)
```
